```python
import random
import re
import pandas as pd
from nltk.chat.util import Chat, reflections


df_hindi = pd.read_csv('new_dataset_hindi.csv')



def recommend_hotels(tags):
    tags_str = ', '.join(tags)
    response = f"बिल्कुल, मैं आपको {tags_str} जैसे होटल सुझा सकता हूँ:"

   
    matches = []
    for idx, row in df_hindi.iterrows():
        row_tags = set(row['Tags'].split(', '))  
        if set(tags).issubset(row_tags):  
            matches.append(row['Hotel_Name'])

    
    if matches:
        unique_hotels = list(set(matches))
        for i, hotel in enumerate(unique_hotels, 1):
            response += f"\n{i}. {hotel}"
    else:
        response += "\nकोई समान होटल नहीं मिला।"

    return response



def respond_recommend_hotels(match):
    if match and match[0]:  
        tags = [tag.strip() for tag in match[0].split(',')]  
        return recommend_hotels(tags) if tags else "कृपया अनुशंसाओं के लिए टैग प्रदान करें।"
    else:
        return "कृपया अनुशंसाओं के लिए टैग प्रदान करें।"


pairs = [
    [
        r"hi|hello|hey|नमस्ते|सुप्रभात",
        ["नमस्ते! मैं आपका होटल सिफारिश चैटबॉट हूँ। मैं आपकी किस प्रकार सहायता कर सकता हूँ?",],
    ],
    [
        r"आपका नाम क्या है",
        ["मेरा नाम चैटबॉट है। आप मुझे चैटबॉट कह सकते हैं।",],
    ],
    [
        r"आप कैसे हैं",
        ["मैं एक मशीन हूँ, इसलिए मैं हमेशा आपकी सहायता के लिए तैयार हूँ। मैं आपकी किस प्रकार सहायता कर सकता हूँ?",],
    ],
    [
        r"अलविदा|विदाई",
        ["अलविदा! यदि आपके पास भविष्य में और प्रश्न हों, तो बेझिझक पूछें। आपका दिन शुभ हो!",],
    ],
    [
        r"(.*) जैसे होटल सुझाएं",  
        [respond_recommend_hotels],
    ],
    [
        r".*",
        ["मुझे खेद है, मैं आपका प्रश्न समझ नहीं पाया। क्या आप कृपया इसे पुनः कह सकते हैं?",],
    ],
]


chat = Chat(pairs, reflections)


def respond(str):
    response = None
    for pattern, responses in chat._pairs:
        match = re.match(pattern, str)

        if match:
            response = random.choice(responses)
            if callable(response):
                response = response(match.groups())
            break

    if response:
        resp = chat._wildcards(response, match)
        return resp

chat.respond = respond
chat.converse()
```
